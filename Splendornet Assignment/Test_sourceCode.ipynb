```python
import pandas as pd

df = pd.read_csv("data-M1aqQE2LtaEELD8QNl-kX.csv")
df
```

```python
new_df = df.drop_duplicates()
# Display the resulting DataFrame without duplicates
print("DataFrame without duplicate records:")
print(new_df)
```

# 1) Find out count of each country from the CSV file. For example, There are 19 records for Country Italy. 


Your Output should look like 

Italy - 19  
India - 35
Srilanka 10
....
etc 


```python
# Find out count of each country from the CSV file. For example, There are 19 records for Country Italy.

country_count = new_df["country"].value_counts()
country_count
```

# 2) Find out sum of numberrange for each county. numberrange is the column given in the CSV file 

Your Output should look like 

Italy - 45678
India - 346757
Srilanka 37643

```python
# Find out sum of numberrange for each county. numberrange is the column given in the CSV file

sum_numberrange = new_df.groupby('country')['numberrange'].sum()
sum_numberrange
```
